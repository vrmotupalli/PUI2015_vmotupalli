```python
import pandas as pd
import urllib2
import json
%pylab inline
import seaborn as sns
import pandas as pd
import numpy as np
import pylab as pl
import zipfile
from scipy import stats
import numpy as np
from sklearn.cluster import KMeans
```

```python
#Got this from asking classmates and looking at Maria's code. I did not think of creating an array of truncated years and
#full years myself. I would not have been able to write this code myself without help.

files = ['94', '95', '96', '97', '98', '99', '00', '01', '02', '03','04', '05', '06', '07', '08', '09', '10', '11', '12', '13']
years = ['1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', 
              '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013']
census_files = []
census_est = {}

for n in files:
    file_path = 'zbp%stotals.zip' % n #runs through all the numbers in files to go through all the zip filenames
    source = zipfile.ZipFile(file_path) #tells zipfile to go to that filepath
    yearcolumn = pd.read_csv(source.open(file_path.replace('.zip','.txt')))
    yearcolumn.columns = [x.upper() for x in yearcolumn.columns]
    yearcolumn = yearcolumn[['ZIP', 'EST']]
    yearcolumn = yearcolumn.rename(columns = {'EST':n})
    census_files.append(yearcolumn) #all the above is just adding columns of years and pulling the data per column year

census_est[0] = census_files[0]

for i in range(1,20):
    census_est[i] = pd.merge(census_files[i], census_est[i-1], on=['ZIP']) 
    
data = census_est[19]
data = data.set_index('ZIP', drop = True)
data = data.reindex_axis([['94', '95', '96', '97', '98', '99', '00', '01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12', '13']], axis=1)
data.columns = years
#data.head()
```

```python

#Loading a file with NYC zipcodes
biggeofile = 'https://nycdatastables.s3.amazonaws.com/2013-08-19T18:18:28.877Z/nyc-zip-code-tabulation-areas-polygons.geojson'
request = urllib2.urlopen(biggeofile)
allzips = json.load(request)

#Pull zip codes and get rid of outside NYC zips
everyzip = [] #collector variable
for i in range(len(allzips['features'])):
    zipcode = allzips['features'][i]['properties']['postalCode'] #stack in the json
    everyzip.append(zipcode)
    
everyzip = map(int, everyzip)

NYCmask = data.index.isin(everyzip)
data = data[NYCmask]
#data.head()
```

```python
#Whiten data by subtracting mean and dividing by stdev

whitened = data
for i in range(len(whitened)):
    whitened.iloc[i] = (whitened.iloc[i]-np.mean(whitened.iloc[i]))/np.std(whitened.iloc[i])
whitened.head()
```

```python
#I'm not entirely sure why I can't just use the df, but it just crashes unless I make it a matrix
whitematrix = whitened.as_matrix()
```

```python
#Plotting the data
pl.figure(figsize=(12, 8))
for i in range(len(whitematrix)):
    pl.plot(years, whitematrix[i], '-', color='blue', linewidth=0.3)
xlim(1994,2013)
ylim(-4,4.25)
pl.xlabel("Year", fontsize=15)
pl.ylabel("Number of Businesses", fontsize=15)
pl.title("Number of Businesses per ZIP code", fontsize=20)
```

```python
#K-means clusters with k=3
k3 = KMeans(n_clusters=3, n_init=100)
k3.fit(whitematrix)
labels3 = k3.labels_

#Cluster centers
kc3_d = np.zeros((3,230))
kc3_m = np.zeros((3,230,20))
kc3 = np.zeros((3,20))

for k in range(3):
    
    for i in range(230):
        if labels3[i] == k:
            kc3_d[k][i] = 1
        else:
            kc3_d[k][i] = 0

    for a in range(20):        
        for b in range(230):
            kc3_m[k][b][a] = kc3_d[k][b]*whitematrix[b][a]

    for i in range(20):
        kc3[k][i] = map(sum, kc3_m[k].transpose())[i]/kc3_d[k].sum()
```

```python
#Plotting 3 clusters        
colormap = ['indianred', 'orange', 'blue', 'green', 'purple']

pl.figure(figsize=(12, 8))
for i in range(len(whitematrix)):
    pl.plot(years, whitematrix[i], '-', color=colormap[labels3[i]], linewidth=0.3, alpha=0.5)
    xlim(1994,2013)
    ylim(-4.5,4.5)
for i in range(3):
    pl.plot(years, kc3[i], '-', color=colormap[i], linewidth=3)

xlim(1994,2013)
ylim(-4,4.25)
pl.xlabel("Year", fontsize=15)
pl.ylabel("Number of Businesses", fontsize=15)
pl.title("K-means Clusters, k = 3", fontsize=20)
```

```python
#K-means clusters with k=4
k4 = KMeans(n_clusters=4, n_init=100)
k4.fit(whitematrix)
labels4 = k4.labels_

#Cluster centers
kc4_d = np.zeros((4,230))
kc4_m = np.zeros((4,230,20))
kc4 = np.zeros((4,20))

for k in range(3):
    
    for i in range(230):
        if labels4[i] == k:
            kc4_d[k][i] = 1
        else:
            kc4_d[k][i] = 0

    for a in range(20):        
        for b in range(230):
            kc4_m[k][b][a] = kc4_d[k][b]*whitematrix[b][a]

    for i in range(20):
        kc4[k][i] = map(sum, kc4_m[k].transpose())[i]/kc4_d[k].sum()
```

```python
#Plotting 4 clusters        
colormap = ['indianred', 'orange', 'blue', 'green', 'purple']

pl.figure(figsize=(12, 8))
for i in range(len(whitematrix)):
    pl.plot(years, whitematrix[i], '-', color=colormap[labels3[i]], linewidth=0.3, alpha=0.5)
    xlim(1994,2013)
    ylim(-4.5,4.5)
for i in range(4):
    pl.plot(years, kc4[i], '.', color=colormap[i], linewidth=3)

xlim(1994,2013)
ylim(-4,4.25)
pl.xlabel("Year", fontsize=15)
pl.ylabel("Number of Businesses", fontsize=15)
pl.title("K-means Clusters, k = 4", fontsize=20)
```

```python

from sklearn.cluster import DBSCAN # just picked one from Google

#DBSCAN 3 clusters
DB = DBSCAN(eps=0.3, min_samples=10).fit(whitematrix)
labelsDB = DB.labels_

#Calculating cluster centers
DB3_d = np.zeros((5,230))
DB3_m = np.zeros((5,230,20))
DB3 = np.zeros((5,20))

for k in range(4):
    
    for i in range(230):
        if labelsDB[i] == k:
            DB3_d[k][i] = 1
        else:
            DB3_d[k][i] = 0

    for a in range(20):        
        for b in range(230):
            DB3_m[k][b][a] = DB3_d[k][b]*whitematrix[b][a]

    for i in range(20):
        DB3[k][i] = map(sum, DB3_m[k].transpose())[i]/DB3_d[k].sum()
```

```python
#Plotting
pl.figure(figsize=(12, 8))
for i in range(len(whitematrix)):
    pl.plot(years, whitematrix[i], 'o', color=colormap[labelsDB[i]], linewidth=0.3)
    xlim(1994,2013)
    ylim(-4.5,4.5)
for i in range(3):
    pl.plot(years, DB3[i], '-', color=colormap[i], linewidth=3)    
    
xlim(1994,2013)
ylim(-4,4.25)
pl.xlabel("Year", fontsize=15)
pl.ylabel("Number of Businesses)", fontsize=15)
pl.title("DBSCAN, 3 clusters", fontsize=20)
```
